```python
import pandas as pd 
import statsmodels.formula.api as smf
```

```python
data=pd.read_csv("Data\Advertising.csv")
data.head()
```

```python
results = smf.ols('sales ~ TV', data=data).fit()
results.summary()
```

```python
results = smf.ols('sales ~ TV + radio + newspaper', data=data).fit()
results.summary()
```

```python
# interaction model
results = smf.ols('sales ~ TV + radio + TV * radio', data=data).fit()
results.summary()
```
